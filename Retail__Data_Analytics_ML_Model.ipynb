#Training Model : ML

```python
data_set.head()
#Messing Data
print(data_set.isna().mean())
md=["MarkDown1","MarkDown2","MarkDown3","MarkDown4","MarkDown5"]
```

```python
#Remplace Missing Data Using Imputer:
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
imputer.fit(data_set[md])
data_set[md]=imputer.transform(data_set[md])
```

```python
data_set.isna().mean()
```

```python
categorical_columns=["Store","Dept","Date"]
non_cat=[i for i in data_set.columns if i != "Weekly_Sales" and i not in categorical_columns]
data_set.loc[:'2012'][non_cat]
```

```python
#Feature Scaling: 
from sklearn.preprocessing import StandardScaler
sscaler=StandardScaler()
data_set.loc[:'2012'][non_cat]=sscaler.fit_transform(data_set.loc[:'2012'][non_cat])
```

```python
data_set.loc[:'2012']
```

```python


x=[i for i in data_set.columns if i != "Weekly_Sales" ]
x=data_set[x].to_numpy()
print(x.shape)
```

```python
#Categorical data encoder
categorical_columns=["Store","Dept","Date"]
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0,1,2])],remainder="passthrough")

x=np.array(ct.fit_transform(x).todense()) #todense
```

```python
#Categorical data encoder
#outher methods
categorical_columns=["Store","Dept","Date"]
from sklearn.preprocessing import OneHotEncoder

onEn=OneHotEncoder()
onEn.fit(data_set[categorical_columns])
en_data=onEn.transform(data_set[categorical_columns]).todense()
pd.DataFrame(en_data,columns=onEn.get_feature_names())
```

```python
x_train=x[:data_set.loc[:'2012'].index.shape[0],:]
x_test=x[data_set.loc[:'2012'].index.shape[0]:,:]
```

```python
print(x_train.shape[0],x_test.shape[0])
```

```python
y_train=data_set.loc[:'2012'].Weekly_Sales.apply(np.log1p).apply(np.log1p).fillna(0).values
y_test=data_set.loc['2012':].Weekly_Sales.apply(np.log1p).apply(np.log1p).fillna(0).values
```
